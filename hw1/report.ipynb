# DINO

```python
import cv2
import mmcv.ops
import mmdet
from mmdet.apis import init_detector, inference_detector
from mmcv.transforms import Compose
from mmengine.utils import track_iter_progress
from mmdet.registry import VISUALIZERS
```

```python
config_file = 'best/dino_config.py'
checkpoint_file = 'best/best.pth'
model = init_detector(config_file, checkpoint_file, device='cuda:0')  # or device='cpu'
visualizer = VISUALIZERS.build(model.cfg.visualizer)

visualizer.dataset_meta = model.dataset_meta
```

```python
# Show the results
img = 'hw1_dataset/test/IMG_2574_jpeg_jpg.rf.ca0c3ad32384309a61e92d9a8bef87b9.jpg'
result = inference_detector(model, img)

img = mmcv.imread(img)
img = mmcv.imconvert(img, 'bgr', 'rgb')


visualizer.add_datasample(
    'result',
    img,
    data_sample=result,
    draw_gt=False,
    show=True)
```

```python
visualizer.show()
```
